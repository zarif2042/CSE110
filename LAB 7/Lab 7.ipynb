```python
#Task 2
def fibonacci(num):
  a=0
  b=1
  print(a,end=" ")
  print(b,end=" ")
  while True:
    c=a+b
    if c>num:
      break
    print(c,end=" ")
    a,b=b,c
fibonacci(10)
print()
fibonacci(5)
```

```python
#Task 4
def up_low(string):
  counter_b=0
  counter_s=0
  for value in string:
    if "a"<=value<="z":
      counter_s+=1
    if "A"<=value<="Z":
      counter_b+=1
  print(f"No. of Uppercase characters :{counter_b}")
  print(f"No. of Lowercase characters :{counter_s}")
up_low("The quick Sand Man")
up_low("HaRRy PotteR")
```

```python
#Task 7
def show_palindrome(num):
  for value in range(1,num+1):
    print(value,end=" ")
  for value1 in range(num-1,0,-1):
    print(value1,end=" ")
show_palindrome(5)
print()
show_palindrome(3)
```

```python
def show_palindrome1(num):
  string = ""
  for value in range(1,num+1):
    string+=str(value)+" "
  for value1 in range(num-1,0,-1):
    string+=str(value1)+" "
  return string
print(show_palindrome1(5))
print(show_palindrome1(3))
```

```python
#Task 9
import math
def area_circumference_generator(radius):
  area = math.pi*radius**2
  circumference = 2*math.pi*radius
  area_circumference = (area,circumference)
  return area_circumference
c_area,c_circum = area_circumference_generator(1)
print(area_circumference_generator(1))
print(f"Area of circle is {c_area} and Circumference is {c_circum}")
```

```python
#Task 10
def make_square(tup):
  start,end= tup
  dic = {}
  for value in range(start,end+1):
    dic[value] = value**2
  return dic
print(make_square((5,9)))
```

```python
#Task 11
def rem_duplicate(tup):
  list_tup = list(tup)
  duplicate_items_r = []
  for value in tup:
    if value not in duplicate_items_r:
      duplicate_items_r.append(value)
  x=tuple(duplicate_items_r)
  return x
print(rem_duplicate(("Hi", 1, 2, 3, 3, "Hi",'a', 'a', [1,2])))
```

```python
#Task 13
def calculator(operator,f_operand,s_operand):
  if operator == "+":
    return f_operand + s_operand
  elif operator == "-":
    return f_operand - s_operand
  elif operator == "*":
    return f_operand * s_operand
  elif operator == "/":
    return f_operand / s_operand
user_operator = input()
user_f_operand = float(input())
user_s_operand = float(input())
print(calculator(user_operator,user_f_operand,user_s_operand))
```

```python
#Task 14
def sentence_word(sentence,position):
  new_sentence = sentence[0]
  remove = ""
  for value in range(1,len(sentence)):
    if value%position != 0:
      new_sentence+=sentence[value]
    else:
      remove+=sentence[value]
  return new_sentence+remove
user_sentence = input()
user_position = int(input())
print(sentence_word(user_sentence,user_position))
```

```python
#Task 15
def cart(order_items,location="Dhanmondi"):
  price = {"Rice":105,"Potato":20,"Chicken":250,"Beef":510,"Oil":85}
  total = 0
  for value in order_items:
    total+=price[value]
  if location == "Dhanmondi":
    total+=30
  else:
    total+=70
  return total
order_items = input("Enter your order items: ").split()
location = input()
if len(location)>0:
  print(cart(order_items,location))
else:
  print(cart(order_items))
```

```python
#Task 16
def splitting_money(amount):
    notes = [500, 100, 50, 20, 10, 5, 2, 1]
    note_counts = []
    for note in notes:
        count = amount // note
        amount %= note
        if count > 0:
            note_counts.append(f"{note} Taka: {int(count)} note(s)")
    return "\n".join(note_counts)

# User input
amount = int(input("Please enter the amount of money: "))
print(splitting_money(amount))
```

```python
#Task 18
def div_check(start, end, divisor1, divisor2):
    total = 0
    for i in range(start, end):
        if (i % divisor1 == 0 or i % divisor2 == 0) and not (i % divisor1 == 0 and i % divisor2 == 0):
            total += i
    return total

# User input
start = int(input("Enter the starting value: "))
end = int(input("Enter the ending value: "))
divisor1 = int(input("Enter the first divisor: "))
divisor2 = int(input("Enter the second divisor: "))

# Function call
print(div_check(start, end, divisor1, divisor2))
```

```python
#Task 21
def individul_bonus_calculation(name, earning, goals, percent):
    bonus = goals * (percent / 100 * earning)
    if goals > 30:
        bonus += 10000
    elif 20 <= goals <= 30:
        bonus += 5000

    print(f"{name} earned a bonus of {int(bonus)} Taka for {goals} goals.")

def cal_bonus(*args):
    for i in range(0, len(args), 4):
        individul_bonus_calculation(args[i], args[i+1], args[i+2], args[i+3])

# Function calls
cal_bonus("Neymar", 1200000, 35, 5)
cal_bonus("Neymar", 1200000, 30, 10, "Jamal", 700000, 19, 5)
cal_bonus("Neymar", 1200000, 35, 5, 'Jamal', 700000, 19, 8, 'Luis', 80000, 25, 10)
```
