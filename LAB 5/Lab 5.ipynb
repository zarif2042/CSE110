```python
#Task 6
input1 = input().split(",")
input1 = [int(value) for value in input1]
max_index = 0
max_num = input1[max_index]
for value in range(len(input1)):
  if input1[value] > max_num:
    max_index = value
    max_num = input1[value]
print(f"Largest number in the list is {max_num} which was found at index {max_index}")
```

```python
#Task 9
nums = []
nums_odd = []
empty_str = ""
num = input()
for value in num:
    if value == " ":
        nums.append(int(empty_str))
        empty_str = ""
    else:
        empty_str+=value
nums.append(int(empty_str))
print(f"Original list:{nums}")
for value1 in nums:
    if value1%2 == 0:
        nums.pop(nums.index(value1))
print(f"Modified list:{nums}")
```

```python
#Task 10
num = input().split(",")
nums = []
nums_m = []
for value in range(len(num)-1):
  nums.append(int(num[value]))
  if int(num[value]) != int(num[value+1]):
    nums_m.append(int(num[value]))
if int(num[-1]) not in nums_m :
  nums_m.append(int(num[-1]))
print(f"Input list:{nums}")
print(f"Modified list:{nums_m}")
```
