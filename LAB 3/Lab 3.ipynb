```python
#Task 28
starting = int(input())
ending = int(input())
count_p=0
count_pr=0

prime_numbers = []
perfect_numbers = []

for k in range(starting,ending+1):
  is_prime=True
  for l in range(2,k):
    if(k%l==0):
      is_prime=False
      break
  if(is_prime==True):
    prime_numbers.append(k)
    count_pr+=1

for m in range(starting,ending+1):
  sum=0
  for n in range(1,m):
    if(m%n==0):
      sum += n
  if(sum==m):
    perfect_numbers.append(m)
    count_p+=1
print(f"Between {starting} and {ending}.")
print(f"Found {count_pr} prime numbers.")
print(f"Found {count_p} perfect number.")
print(f"Prime numbers:",end="")
for value in prime_numbers:
    if(value == prime_numbers[-1]):
        print(value,end="")
    else:
        print(value,end=",")
print(f"\nPerfect numbers:",end="")
for value in perfect_numbers:
    if(value == perfect_numbers[-1]):
        print(value,end="")
    else:
        print(value,end=",")
```

```python
#Task28(alternative)
start = int(input())
end = int(input())

prime_numbers = []
perfect_numbers = []

for num in range(start, end + 1):
    # checking for prime numbers
    if num > 1:
        for i in range(2, num):
            if (num % i) == 0:
                break
        else:
            prime_numbers.append(str(num))

    # checking for perfect numbers
    sum = 0
    for i in range(1, num):
        if num % i == 0:
            sum += i
    if sum == num:
        perfect_numbers.append(str(num))

print(f"Between {start} and {end},")
print(f"Found {len(prime_numbers)} prime numbers")
print(f"Found {len(perfect_numbers)} perfect numbers")
print("Prime numbers: " + ", ".join(prime_numbers))
print("Perfect numbers: " + ", ".join(perfect_numbers))
```

```python
#Task 29
start = int(input())
end = int(input())
n = int(input())

for i in range(start, end+1):
    product = 1
    temp = i
    while temp > 0:
        product *= temp % 10
        temp //= 10
    if product % n == 0:
        print(product, end=" ")
```
