```python
#Task10
string = input()
index1 = 0
index2 = 0
string1 = ""
for value in string:
  if(value == ','):
    break
  else:
    index2 = index2 + 1
index3 = index2+2
index4 = len(string)
while index1<index2 and index3<index4:
  string1 += string[index1]+string[index3]
  index1+=1
  index3+=1
if(index1<index2):
  string1 += string[index1:index2]
else:
  string1 += string[index3:index4]
print(string1)
```

```python
#Task10(Alternative)
# take input from user
user_input = input()

# split the input into two strings
str1, str2 = user_input.split(', ')

# find the minimum length
min_len = min(len(str1), len(str2))

# create the mixed string with alternative characters from each string
result = ''
for i in range(min_len):
    result += str1[i]
    result += str2[i]

# append the leftover chars at the end
result += str1[min_len:]
result += str2[min_len:]

print(result)
```

```python
#Task19
input_string = input("Enter a string: ")
result = []
uppercase = True

for char in input_string:
      if char.isalpha():
          if uppercase:
              result.append(char.upper())
          else:
              result.append(char.lower())
          uppercase = not uppercase
      else:
          result.append(char)

for value in result:
    print(value,end="")
```
