```python
#Task 22
canvas = int(input("Enter the number of canvas:"))
paint = int(input("Enter the number of paint tubes:"))

total = canvas*120 + paint*75

print(f"Previous total: {total}")

if(total>=0 and total<=299):
  discount=0
elif(total>=300 and total<=499):
  discount = 10
elif(total>=500 and total<=749):
  discount = 20
elif(total>=750 and total<=999):
  discount = 50
elif(total>=1000):
  discount = 150

print(f"New total after discount: {total-discount}")
```
