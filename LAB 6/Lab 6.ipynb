```python
#Task 19
from collections import defaultdict

def group_tuples(tuples):
    result = defaultdict(list)
    for tuple in tuples:
        result[tuple[1]].append(tuple)
    return dict(result)

# Test with provided examples
tuples1 = [(20, 80), (31, 80), (1, 22), (88, 11), (27, 11)]
print(group_tuples(tuples1))
# Output: {80: [(20, 80), (31, 80)], 22: [(1, 22)], 11: [(88, 11), (27, 11)]}

tuples2 = [(20, 'Sad'), (31, 'Sad'), (88, 'NotSad'), (27, 'NotSad')]
print(group_tuples(tuples2))
# Output: {'Sad': [(20, 'Sad'), (31, 'Sad')], 'NotSad': [(88, 'NotSad'), (27, 'NotSad')]}
```
